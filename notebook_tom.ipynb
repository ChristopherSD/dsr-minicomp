```python
def one_hot_encoder_fit_transform(df, col_name):
    
    enc = OneHotEncoder(handle_unknown='ignore', sparse=False)
    
    enc.fit(df[col_name].values.reshape(-1, 1))
    
    return one_hot_encoder_transform(df, col_name, enc), enc

def one_hot_encoder_transform(df, col_name, enc):
    
    encoded_column = pd.DataFrame(enc.transform(df[col_name].values.reshape(-1, 1)),
                                  columns=[col_name + '_' + str(item) for item in range(len(enc.categories_[0]))],
                                  index=df.index
                                  
                                 )
    return pd.concat([df, encoded_column], axis=1).drop(col_name, axis=1)

def target_encoder(df, x_col, y_col):
    
    agg = df.groupby(x_col)[y_col].agg(np.mean).to_dict()
    
    return df[x_col].apply(lambda x: agg[x]).values
```

```python
col_to_log_transform = ['Customers', 'CompetitionDistance', 'CompetitionSince', 'Promo2SinceNWeeks']
col_to_str = ['Promo', 'StateHoliday', 'SchoolHoliday', 'StoreType', 'Assortment', 'Promo2']

target_var = 'Sales'

train_vars = [
    'DayOfWeek',
    #'Customers',
    'Promo',
    'StateHoliday',
    'SchoolHoliday',
    'StoreType',
    'Assortment',
    #'CompetitionDistance',
    'Promo2',
    #'CompetitionSince',
    'Competition_missing',
    #'Promo2SinceNWeeks',
    'Promo2SinceNWeeks_missing',
    'PromoStarted',
    'Store',
    #'Week',
    'sin_week',
    'cos_week',
    #'Month',
    'sin_month',
    'cos_month',
    'Customers_log', 
    'CompetitionDistance_log', 
    'CompetitionSince_log', 
    'Promo2SinceNWeeks_log'
]

enc_var_list = [
    'DayOfWeek',
    'Promo',
    'StateHoliday',
    'SchoolHoliday',
    'StoreType',
    'Assortment',##consider labeling
    'Promo2'
]

def custom_transformer(df):

    # creating new features
    generate_CompetitionSince(df)
    
    generate_Promo2SinceNWeeks(df)
    
    generate_PromoStarted(df)
    
    df['Month'] = df['Date'].dt.month
    df['Week'] = df['Date'].dt.week

    # log transform    
    for col in col_to_log_transform:
        df[col + '_log'] = log_transform(df[col])

    #sin-cos transform
    sin_month, cos_month = generate_cyclic_feature_month(df)
    df['sin_month'] = sin_month
    df['cos_month'] = cos_month
    
    sin_week, cos_week = generate_cyclic_feature_week(df)
    df['sin_week'] = sin_week
    df['cos_week'] = cos_week

    # replace and change of types
    df['Promo'] = df['Promo'].replace({0.0: 'n', 0: 'n', 1.0: 'y', 1: 'y'}).astype(str)
    df['Promo2'] = df['Promo2'].replace({0.0: 'n', 0: 'n', 1.0: 'y', 1: 'y'}).astype(str)
    df['SchoolHoliday'] = df['SchoolHoliday'].replace({0.0: 'n', 0: 'n', 1.0: 'y', 1: 'y'}).astype(str)
    df['StateHoliday'] = df['StateHoliday'].replace({'0': 'n'}).astype(str)

    for col in col_to_str:
        df[col] = df[col].astype(str)
```

```python
def metric(preds, actuals):
    preds = preds.reshape(-1)
    actuals = actuals.reshape(-1)
    assert preds.shape == actuals.shape
    return 100 * np.linalg.norm((actuals - preds) / actuals) / np.sqrt(preds.shape[0])

def rmpse(actuals, preds):
    preds = np.array(preds)
    actuals = np.array(actuals)
    preds = preds.reshape(-1)
    actuals = actuals.reshape(-1)
    assert preds.shape == actuals.shape
    return 100 * np.linalg.norm((actuals - preds) / actuals) / np.sqrt(preds.shape[0])
```

```python
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
import math
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import make_scorer

import pickle


df_base = data_imputed.copy()

kf = KFold(n_splits = 5, shuffle=True, random_state=42)

for i, (train_index, test_index) in enumerate(kf.split(data_imputed)):
    
    if i < 4:
        continue
    
    X_train = df_base.iloc[train_index]
    y_train = df_base.iloc[train_index][target_var]
    
    X_test = df_base.iloc[test_index]
    y_test = df_base.iloc[test_index][target_var]
    
    # Transform
    custom_transformer(X_train)
    custom_transformer(X_test)

    # target encoding
    new_col, target_encoder = target_encode_Stores(X_train)
    X_train['Store'] = new_col
    
    new_col, _ = target_encode_Stores(X_test, target_encoder)
    X_test['Store'] = new_col
    
    # Keep selected features
    X_train = X_train[train_vars]
    X_test = X_test[train_vars]
    
    # OneHotEncoding
    for col in enc_var_list:
        X_train, one_hot_encoder = one_hot_encoder_fit_transform(X_train, col)
        X_test = one_hot_encoder_transform(X_test, col, one_hot_encoder)
    
    # Training
    rfc = RandomForestRegressor(n_jobs=-1, random_state=42)
    
    param_grid = {'max_features': ['auto'], #['auto', math.floor(math.sqrt(X_train.shape[1]))]
                  'n_estimators': [300], #[50, 100, 200, 500]
                  'max_depth': [75] #[20, 50, 75, 100]
                 }
    
    rmpse_scorer = make_scorer(rmpse, greater_is_better=False)

    grid2 = GridSearchCV(estimator=rfc, param_grid=param_grid, scoring=rmpse_scorer)
    grid2.fit(X_train, y_train)
    
    y_pred = grid2.best_estimator_.predict(X_test)
    print(metric(y_pred, np.array(y_test)))
    
    pkl_filename = "./models/model_tom_rf_rmpse.pkl"
    with open(pkl_filename, 'wb') as file:
        pickle.dump(grid2.best_estimator_, file)
    
```
